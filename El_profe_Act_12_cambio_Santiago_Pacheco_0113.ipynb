## Las variables
1. CP --> Centro del Producto
2. CD --> Cantiadad de Dinero
3. Cambio
## Formulas
* $ Cambio = CD - CP $
## Diagrama de flujo lucidchart
<img src="https://raw.githubusercontent.com/santiagopacheco128/Calcular-Cambio-if-else/refs/heads/main/DF%20if%20else%20cambio.png" width="350" height="500">



```python
# Para cuando es verdadero (Si)
# Calculando cambio usando if else
# Datos de entrada CD, CP
# Mostrar cambio si es falso
# Si es verdadero mostrar no alcanza
CP = int(input("Ingrese el costo del producto: "))
CD = int(input("Ingrese la cantidad de dinero: "))
if CP > CD:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
print("Fin del programa")
print("Programa realizado por Santiago Pacheco NL = 43")
```

```python
# Para cuando es falso (No)
# Calculando cambio usando if else
# Datos de entrada CD, CP
# Mostrar cambio si es falso
# Si es verdadero mostrar no alcanza
CP = int(input("Ingrese el costo del producto: "))
CD = int(input("Ingrese la cantidad de dinero: "))
if CP > CD:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
print("Fin del programa")
print("Programa realizado por Santiago Pacheco NL = 43")
```
